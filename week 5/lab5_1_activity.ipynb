รูปแบบ: กิจกรรมระดมความคิด (กลุ่มย่อย หรือ ทั้งชั้นเรียน)

คำชี้แจง: ให้นักเรียนช่วยกันยกตัวอย่างสถานการณ์หรือปัญหาในชีวิตจริงที่ Dictionary และ Set เหมาะสมจะถูกนำมาใช้เป็นโครงสร้างข้อมูล พร้อมอธิบายเหตุผลประกอบ

คำถามกระตุ้นความคิด (Prompts):

```python
{"hp": 100, "mp": 50, "atk": 20, "def": 15}
```

= ค่าพลังตัวละคร	แต่ละค่ามีความหมายต่างกัน การอ้างอิงด้วยชื่อ (key) อ่านเข้าใจง่ายกว่าอ้างอิงด้วยตำแหน่ง index

Dictionary จะช่วยจัดเก็บข้อมูลสินค้าในร้านค้าออนไลน์ได้อย่างไร? (เช่น ข้อมูลรายละเอียดสินค้าโดยใช้อ้างอิงจากรหัสสินค้า/ID)

```python
products = {
      "SKU1001": {"name": "เสื้อยืด", "price": 250, "stock": 30},
      "SKU1002": {"name": "กางเกงยีนส์", "price": 690, "stock": 12}
}
```

= เวลาลูกค้าสั่งซื้อสินค้ารหัส SKU1001 ระบบสามารถดึงราคาและจำนวนสต็อกได้ทันทีโดยไม่ต้องค้นหาทีละรายการในลิสต์สินค้าทั้งหมด (เร็วกว่ามากเมื่อร้านมีสินค้าเป็นหมื่นรายการ)

Set จะมีประโยชน์อย่างไรในการจัดการสิทธิ์ของผู้ใช้ หรือการจัดการแท็ก? (เช่น บทบาทสิทธิ์ใช้งานที่ไม่ซ้ำกัน, แฮชแท็กที่ไม่ซ้ำกัน

```python
user_permissions = {"read", "write", "delete"}
```

= สิทธิ์แต่ละอย่างไม่ควรซ้ำกัน (จะมีสิทธิ์ "write" สองครั้งไปทำไม) และการเช็กว่า "delete" in user_permissions ทำได้เร็วมาก เหมาะกับระบบที่ต้องเช็กสิทธิ์บ่อยๆ เช่น ทุกครั้งที่มีการกดปุ่มลบ

นึกออกไหมว่ามีสถานการณ์ไหนบ้างที่เราต้องเช็กอย่างรวดเร็วว่า "มีข้อมูลนี้อยู่ในกลุ่มข้อมูลขนาดใหญ่หรือไม่" โดยไม่สนใจลำดับก่อนหลัง? (เช่น รายชื่อ IP Address ที่ถูกบล็อก, การนับจำนวนผู้เข้าชมเว็บที่ไม่นับคนซ้ำ)
