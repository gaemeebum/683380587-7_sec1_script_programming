#Block A : Evaluation Check
temp = 25
is_raining = True

if temp > 20 and not is_raining:
    print("Perfect day for a picnic!")
elif temp <= 20 or is_raining:
    print("Stay indoors or bring an umbrella.")
else:
    print("Unsure about the weather.")

ลำดับการทำงาน
1. กำหนด temp = 25
2. กำหนด is_raining = True
3. ตรวจสอบเงื่อนไขแรก
temp > 20 and not is_raining

(temp > 20 → 25 > 20 → True
not is_raining → not True → False
True and False → False)
จึงไม่เข้า if

4. ตรวจสอบ elif
temp <= 20 or is_raining

(temp <= 20 → 25 <= 20 → False
is_raining → True
False or True → True)

output
Stay indoors or bring an umbrella.
