Activity 3.2: Pattern Printing with Nested Loops

```python
# A square of asterisks:
rows = 5
cols = 5

for i in range(rows):
    for j in range(cols):
        print("*", end="")
    print()
```

```python
# A right-angled triangle of asterisks:
rows = 5

for i in range(1, rows + 1):
    for j in range(i):
        print("*", end="")
    print()
```

Challenge: An inverted right-angled triangle or a pyramid.
Purpose: Reinforce understanding of nested loops and the relationship between loop variables and printed output. Encourages algorithmic thinking and problem-solving.


```python
# inverted right-angled triangle
rows = 5

for i in range(rows, 0, -1):
    for j in range(i):
        print("*", end="")
    print()
```

```python
# a pyramid
rows = 5

for i in range(1, rows + 1):
    for j in range(rows - i):
        print(" ", end="")
    for k in range(2 * i - 1):
        print("*", end="")

    print()
```
